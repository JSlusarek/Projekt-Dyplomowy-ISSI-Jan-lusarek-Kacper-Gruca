```python
def simulate_dishwasher_usage(mode: str, cycles: int = 1):
    """
    Simulates electricity usage for a dishwasher based on the selected mode and number of cycles.

    Parameters:
        mode (str): One of 'eco', 'normal', 'fast'.
        cycles (int): Number of dishwasher cycles to simulate.

    Returns:
        dict: Dictionary containing mode, energy used (kWh), duration (minutes), and total stats.
    """

    # Define usage profiles
    modes = {
        "eco": {"energy_kwh": 0.54, "duration_min": 180},
        "normal": {"energy_kwh": 1.0, "duration_min": 120},
        "fast": {"energy_kwh": 1.5, "duration_min": 60},
    }

    # Validate mode
    if mode not in modes:
        raise ValueError("Invalid mode. Choose from 'eco', 'normal', or 'fast'.")

    # Calculate usage
    energy = modes[mode]["energy_kwh"] * cycles
    duration = modes[mode]["duration_min"] * cycles

    return {
        "mode": mode,
        "cycles": cycles,
        "total_energy_kwh": energy,
        "total_duration_min": duration,
        "per_cycle": modes[mode],
    }
```

```python
def simulate_boiling_water(appliance: str, liters: float = 1.0):
    """
    Simulates the energy usage and time required to boil a given amount of water using either a kettle or induction hob.

    Parameters:
        appliance (str): 'kettle' or 'induction'.
        liters (float): Amount of water in liters.

    Returns:
        dict: Appliance used, energy consumed (kWh), time taken (minutes), and assumptions.
    """
    # Constants
    specific_heat_kj_per_kg_c = 4.186  # kJ/kg°C
    temp_rise_c = 80  # From room temperature (~20°C) to boiling (~100°C)
    energy_kj = liters * specific_heat_kj_per_kg_c * temp_rise_c
    energy_kwh_theoretical = energy_kj / 3600  # Convert to kWh

    # Appliance efficiencies and heating power
    appliances = {
        "kettle": {"efficiency": 0.90, "power_kw": 2.0},       # 2kW kettle
        "induction": {"efficiency": 0.74, "power_kw": 1.8}     # 1.8kW induction hob
    }

    if appliance not in appliances:
        raise ValueError("Invalid appliance. Choose 'kettle' or 'induction'.")

    efficiency = appliances[appliance]["efficiency"]
    power_kw = appliances[appliance]["power_kw"]

    # Actual energy considering efficiency
    energy_kwh_actual = energy_kwh_theoretical / efficiency

    # Time in hours = energy / power; convert to minutes
    time_minutes = (energy_kwh_actual / power_kw) * 60

    return {
        "appliance": appliance,
        "water_liters": liters,
        "energy_kwh": round(energy_kwh_actual, 3),
        "time_minutes": round(time_minutes, 1),
        "assumptions": {
            "efficiency": efficiency,
            "power_kw": power_kw,
            "temperature_rise_c": temp_rise_c,
            "specific_heat_kj_per_kg_c": specific_heat_kj_per_kg_c
        }
    }
```

```python
class Kettle:
    def __init__(self, appliance: str = 'kettle'):
        """
        Inicjalizacja obiektu Kettle.

        Parametry:
            appliance (str): 'kettle' lub 'induction'.
        """
        self.appliance = appliance.lower()
        self.appliances = {
            "kettle": {"efficiency": 0.90, "power_kw": 2.0},       # typowy czajnik 2kW
            "induction": {"efficiency": 0.74, "power_kw": 1.8}     # typowa płyta indukcyjna
        }

        if self.appliance not in self.appliances:
            raise ValueError("Invalid appliance. Choose 'kettle' or 'induction'.")

        self.efficiency = self.appliances[self.appliance]["efficiency"]
        self.power_kw = self.appliances[self.appliance]["power_kw"]
        self.temp_rise_c = 80  # domyślnie: z 20°C do 100°C
        self.specific_heat_kj_per_kg_c = 4.186  # kJ/kg°C

    def boil(self, liters: float = 1.0) -> dict:
        """
        Symulacja gotowania wody.

        Parametry:
            liters (float): Ilość wody w litrach.

        Zwraca:
            dict: Szczegóły zużycia energii i czasu.
        """
        # Energia potrzebna teoretycznie
        energy_kj = liters * self.specific_heat_kj_per_kg_c * self.temp_rise_c
        energy_kwh_theoretical = energy_kj / 3600  # konwersja do kWh

        # Rzeczywiste zużycie energii z uwzględnieniem efektywności
        energy_kwh_actual = energy_kwh_theoretical / self.efficiency

        # Czas gotowania w minutach
        time_minutes = (energy_kwh_actual / self.power_kw) * 60

        return {
            "appliance": self.appliance,
            "water_liters": liters,
            "energy_kwh": round(energy_kwh_actual, 3),
            "time_minutes": round(time_minutes, 1),
            "assumptions": {
                "efficiency": self.efficiency,
                "power_kw": self.power_kw,
                "temperature_rise_c": self.temp_rise_c,
                "specific_heat_kj_per_kg_c": self.specific_heat_kj_per_kg_c
            }
        }
```

```python
def simulate_boiling_water_with_costs(appliance: str, liters: float = 1.0):
    """
    Simulates boiling water using kettle, induction or gas, including energy usage and cost.

    Parameters:
        appliance (str): 'kettle', 'induction', or 'gas'.
        liters (float): Amount of water in liters.

    Returns:
        dict: Appliance used, energy consumed (kWh), cost (PLN), time (min), and assumptions.
    """

    # Ceny energii i gazu (możesz zmienić na aktualne wartości)
    electricity_price_per_kwh = 0.85  # zł/kWh
    gas_price_per_m3 = 2.80           # zł/m³ (łącznie z dystrybucją)

    # Stałe fizyczne
    specific_heat_kj_per_kg_c = 4.186
    temp_rise_c = 80
    energy_kj = liters * specific_heat_kj_per_kg_c * temp_rise_c
    energy_kwh_theoretical = energy_kj / 3600

    # Parametry urządzeń
    appliances = {
        "kettle": {"efficiency": 0.90, "power_kw": 2.0},
        "induction": {"efficiency": 0.74, "power_kw": 1.8},
        "gas": {
            "efficiency": 0.60,
            "energy_per_liter_m3": 0.015,
            "gas_energy_density_kwh_per_m3": 9.5,
            "boiling_time_minutes": 6.0
        }
    }

    if appliance not in appliances:
        raise ValueError("Choose 'kettle', 'induction', or 'gas'.")

    if appliance == "gas":
        m3_gas = liters * appliances["gas"]["energy_per_liter_m3"]
        energy_kwh = m3_gas * appliances["gas"]["gas_energy_density_kwh_per_m3"] / appliances["gas"]["efficiency"]
        time_minutes = appliances["gas"]["boiling_time_minutes"] * liters
        cost_pln = m3_gas * gas_price_per_m3
        return {
            "appliance": appliance,
            "water_liters": liters,
            "energy_kwh": round(energy_kwh, 3),
            "cost_pln": round(cost_pln, 2),
            "time_minutes": round(time_minutes, 1),
            "assumptions": {
                "efficiency": appliances["gas"]["efficiency"],
                "gas_usage_m3": round(m3_gas, 4),
                "gas_energy_density_kwh_per_m3": appliances["gas"]["gas_energy_density_kwh_per_m3"],
                "gas_price_per_m3": gas_price_per_m3
            }
        }
    else:
        efficiency = appliances[appliance]["efficiency"]
        power_kw = appliances[appliance]["power_kw"]
        energy_kwh_actual = energy_kwh_theoretical / efficiency
        time_minutes = (energy_kwh_actual / power_kw) * 60
        cost_pln = energy_kwh_actual * electricity_price_per_kwh

        return {
            "appliance": appliance,
            "water_liters": liters,
            "energy_kwh": round(energy_kwh_actual, 3),
            "cost_pln": round(cost_pln, 2),
            "time_minutes": round(time_minutes, 1),
            "assumptions": {
                "efficiency": efficiency,
                "power_kw": power_kw,
                "temperature_rise_c": temp_rise_c,
                "specific_heat_kj_per_kg_c": specific_heat_kj_per_kg_c,
                "electricity_price_per_kwh": electricity_price_per_kwh
            }
        }
```

```python
print(simulate_boiling_water_with_costs("kettle", 1.0))
print(simulate_boiling_water_with_costs("induction", 1.0))
print(simulate_boiling_water_with_costs("gas", 1.0))
```
